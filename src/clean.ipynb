```python
cast['personaje']= cast['personaje'].str.replace('…', '')
cast['personaje']= cast['personaje'].str.replace(r'(voz)', '')

cast['personaje']= cast['personaje'].str.replace('Dr.', 'Doctor')
cast['personaje']= cast['personaje'].str.replace('Captain America', 'Steve Rogers')

def limpiar_personaje(x):
    
    if 'Maria Hill' in x:
        return 'Maria Hill'
    elif 'Coulson' in x:
        return 'Agent Coulson'
    elif 'Rhodes' in x:
        return 'James Rhodes'
    elif 'Barnes' in x:
        return 'James Barnes'
    elif 'Strange' in x:
        return 'Stephen Strange'
    elif 'Shaun' in x:
        return 'Shang-Chi'
    elif 'Michelle' in x:
        return 'MJ'
    elif 'Ross' in x and 'Everett' not in x:
        return 'Secretary Ross'
    elif 'Groot' in x:
        return 'Groot'
    elif 'Flash' in x:
        return 'Flash Thompson'
    elif 'jarvis' in x.lower():
        return 'Jarvis'
    elif 'Yondu' in x:
        return 'Yondu Udonta'
    else:
        return x.strip()
```

```python
cast['personaje'] = cast['personaje'].apply(limpiar_personaje)
```

```python
names['Character']= names['Character'].str.replace('T’Challa', "T'Challa")
names['Character']= names['Character'].str.replace('Steve Rodgers', "Steve Rogers")
```

```python
def poner_heroe(x, y):
    
    if type(x) != str:
        if y == 'Scott Lang':
            return 'Ant-Man'
        elif y == 'Thor':
            return 'Thor'
        elif y == 'Sam Wilson':
            return 'Falcon'
        elif y =='Shang-Chi':
            return 'Shang-Chi'
        elif  y =='Sersi' in x:
            return 'Sersi'
        else:
            return x
    else:
        return x
            
    
```
